```python
import numpy as np
import pandas as pd
```

## Series from lists

```python
# string

fruits= ['apple', 'guava','lichi','mango','graps','orange','banana']
pd.Series(fruits)
```

```python
# integers
marks= [60, 55, 49, 70, 66]
pd.Series(marks)
```

```python
# custom index
marks= [60, 55, 49, 70, 66]
subjects= ['physics','chemistry','biology','math','ict']
pd.Series(marks, index= subjects)
```

```python
marks= pd.Series(marks, index= subjects, name="Jobayer's marks.")
marks 
```

## Series from Dictionary

```python
marks={
    'physics': 60,
    'chemistry': 67,
    'biology' :70,
    'math': 80,
    'ict': 78
}

mark_series=pd.Series(marks,name="Jobayer's marks." )
mark_series
```
