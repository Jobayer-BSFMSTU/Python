```python
import numpy as np
import pandas as pd
```

```python
movies= pd.read_csv('imdb-top-1000.csv')
movies
```

```python
genre= movies.groupby('Genre')
genre
```

## group operation

```python
genre.min()
```

## find the 3 highest gross

```python
movies.groupby('Genre')['Gross'].sum().sort_values(ascending= False).head(3)
```

## find highest IMDB_Rating

```python
movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending= False)
```

## find the best director

```python
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending= False)
```

## find number of movies done by actors

```python
movies.groupby('Star1')['Series_Title'].count().sort_values(ascending= False)
```

## find total number of groups

```python
len(movies.groupby('Genre'))
```

## find total item in each group

```python
movies.groupby('Genre').size().sort_values(ascending = False)
```

## first()/ last()/ nth item

```python
genre= movies.groupby('Genre')
genre.first()
```

```python
genre.last()
```

```python
genre.nth(6)
```

## get_group  /  filtering

```python
genre.get_group('Adventure').sort_values('Runtime',ascending= False)
```

```python
movies[movies['Genre'] == 'Adventure'].sort_values('Runtime',ascending= False)
```

## agg methods(list/ dict)

```python
genre.agg({
    'Released_Year': 'max',
    'IMDB_Rating': 'mean',
    'Gross':'sum',
    'No_of_Votes':'sum'
})
```

```python
genre.agg(['min', 'max', 'sum'])
```

## looping on group

```python
for group,data in genre:
    print(data)
```

## find the highes rated movie using loop

```python
for group,data in genre:
    print( data[data['IMDB_Rating'] == data['IMDB_Rating'].max()])
```

```python
df_list = []

for group, data in movies.groupby('Genre'):
    max_row = data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]
    df_list.append(max_row)

df = pd.concat(df_list, ignore_index=True)
df
```

## use of apply in DataFrame

```python
def mov(group):
    group['Series_Title'].str.startswith('A').sum()
    return group
```

```python
genre.apply(mov)
```

## finding rank 

```python
def rank_mov(group):
    group['genre_rank'] = group['IMDB_Rating'].sort_values(ascending = True)
    return group
```

```python
genre.apply(rank_mov)
```

```python
ipl= pd.read_csv('deliveries.csv')
ipl
```

```python
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending = False)
```

```python
six= ipl[ipl['batsman_runs'] == 6]
six.groupby('batsman')['batsman'].count().sort_values(ascending = False).head(1).index[0]
```

```python
virat = ipl[ipl['batsman']=='V Kohli']
virat.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending = False).reset_index()
```

## create a function that can return the highest score of any batsman

```python
def most_runs(batsman):
    temp_df = ipl[ipl['batsman'] == batsman]
    return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending = False).head(1)

```

```python
most_runs('S Dhawan')
```

```python
most_runs('DA Warner')
```
